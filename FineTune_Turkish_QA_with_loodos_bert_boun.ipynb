```
import transformers
transformers.logging.set_verbosity_error()
from transformers import AutoTokenizer, AutoModelForQuestionAnswering
import torch

# Load the fine-tuned model
tokenizer = AutoTokenizer.from_pretrained("/content/drive/MyDrive/TURKISHLLM/modelloodosboun/")
model = AutoModelForQuestionAnswering.from_pretrained("/content/drive/MyDrive/TURKISHLLM/modelloodosboun/")
model.push_to_hub("muratsimsek003/turkish-loodos-bert-base-uncased-boun-qa")
tokenizer.push_to_hub("muratsimsek003/turkish-loodos-bert-base-uncased-boun-qa")
```

```
from transformers import pipeline
question_answerer = pipeline("question-answering",model = model,tokenizer = tokenizer)
```
