```python
from transformers import pipeline

# Modeli pipeline ile yükleme
generator = pipeline('text-generation', model='muratsimsek003/turkish_llama3_finetuned')

# Metin üretimi
text = "Ankara'nın tarihini anlat"
result = generator(text, max_length=50, num_return_sequences=1)

# Üretilen metni yazdırma
print(result[0]['generated_text'])
```
